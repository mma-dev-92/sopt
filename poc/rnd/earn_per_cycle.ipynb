### Earn per Cycle

Assumptions / simplifications:
- storage technical parameters are constant (will not get any worse)
- storage CAPEX was paid upfront (no capital costs)
- storage OPEX is constant in each year
- storage lifetime is Y years
- storage in its lifetime will perform exactly $N$ cycles

With that simplifications, the $NVP$ with the desired yearly rate of 
return $r$ is given by
$$
NPV = -\hbox{CAPEX} + \sum_{y=1}^Y\frac{CF_y}{(1 + r)^y}.
$$
Total cash flow $CF_y$ in a year $y$ can be simplified as
$$
CF_y = -\hbox{OPEX} + \hbox{revenue}_y
$$
Let's assume, that in each year the revenue is the same, then our formula 
for $NPV$ can be simplified as
$$
NPV = -\hbox{CAPEX} + (\hbox{revenue} - OPEX)\cdot\sum_{y=1}^Y\frac{1}{(1+r)^y}.
$$
Now, we are looking the minimal value of the revenue, for which the $NPV > 0$. 
Let's assume $r=wacc$. Then, we are looking at the inequality
$$
-\hbox{CAPEX} + (\hbox{revenue} - OPEX)\cdot\sum_{y=1}^Y\frac{1}{(1+wacc)^y} 
\ge 0
$$
We can easily solve this inequality
$$
\hbox{revenue} \ge \frac{\hbox{CAPEX}}{\sum_{y=1}^Y\frac{1}{(1+wacc)^y}} + OPEX
$$
Therefore, in that simplified scenario we see, that the mean revenue per 
each year can be calculated as
$$
\hbox{revenue} = \frac{\hbox{CAPEX}}{\sum_{y=1}^Y\frac{1}{(1+wacc)^y}} + OPEX.
$$

If we assume, that for each year the number of cycles will be the same, 
our simplified version of $epc$ is
$$
epc = \frac{\hbox{revenue}\cdot Y}{N}
$$
Which in final form is the formula
$$
epc = \frac{Y}{N}\cdot \left(\frac{\hbox{CAPEX}}{\sum_{y=1}^Y\frac{1}{(1+wacc)^y}} + OPEX\right)
$$

### Expected Earn per Energy Unit ($eee$) 

The idea is as follows. Each energy storage have limited amount of cycles it can 
perform (load/reload cycles). The idea is to estimate the minimal mean earn per 
cycle ($epc$), for which the $NPV > 0$. 

If in the given moment buy/sell operation would have less profit, than the 
$epc - \epsilon$, then the storage should do nothing (otherwise that would be 
inefficient usage of the cycle), risking $NPC < 0$.

To include this logic in the model, we introduce the earn per cycle 
parameter ($epc > 0$). 

Next we can estimate the energy, that we need to buy from the energy market for 
a single cycle. We assume, that in the first moment of the cycle $t_0$, we have
$soc_{t_0} = cap\cdot dod_{min}$, and the max value of the $soc$ during the cycle
equals $cap\cdot dod_{max}$. Therefore, the total energy that needs to be bought
from the market for a single cycle is $\frac{cap\cdot (dod_{max} - dod_{min})}{\eta_{load}}$.

It means, that to be sure, that all cycles will give at least $epc$ in revenue,
we can compute an expected earn by (energy) unit ($\hbox{eeu}$),
that will tell us, how much we need to earn per each energy unit that we bought

$$
\hbox{eeu} = \frac{cap}{epc}\cdot \frac{dod_{max} - dod_{min}}{\eta_{load}}.
$$

To modify the obj, we need to

$$
obj \leftarrow obj -\left(\hbox{eeu} \cdot \sum_{t\in T}load_t\right)
$$
